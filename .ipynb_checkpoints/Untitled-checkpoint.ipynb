```python
from scipy.stats import chi2_contingency as cc   
for i in range(0,6):
    d=pd.crosstab(obj.iloc[:,i],df['Loan_Status'],margins=False)
    print(d)
    stats,p,dof,expect=cc(d)
    print('\n')
    if p <=0.05:
        print('reject H0 of',obj.columns[i])
    else:
        print('fail to reject H0 of',obj.columns[i])
    print('\n')
```

```python
from statsmodels.stats import weightstats as zte
j=0
for i in s.iloc[:,:-1]:
    z,p=zte.ztest(df[df['Loan_Status']=='Y'][i],x2=df[df['Loan_Status']=='N'][i],value=df[i].mean())
    print('p value of {0} : {1}'.format(s.columns[j],p))
    if p <= 0.05:
        print('reject H0 of',s.columns[j])
    else:
        print('fail to reject of',s.columns[j])
    j=j+1
    print('\n')
```

```python
df=df[['ApplicantIncome','CoapplicantIncome','LoanAmount','Married','Education','Property_Area','Loan_Status']]
```

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
for i in range(3,7):
    df.iloc[:,i]=le.fit_transform(df.iloc[:,i])
```

```python
from sklearn.model_selection import train_test_split
train,test=train_test_split(df,test_size=0.2,shuffle=True,random_state=42)
```

```python
from sklearn.feature_selection import RFECV
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from xgboost import XGBClassifier
```

### Logistic Regression 

```python
lr=LogisticRegression(random_state=0)
select=RFECV(lr,step=1,cv=5).fit(train.iloc[:,:-1],train.iloc[:,-1])
```

```python
pd.DataFrame({'Name of column':train.columns[:-1],'Support':select.support_,'Ranking':select.ranking_})
```

### Random Forest

```python
X,Y=train.iloc[:,:-1],train.iloc[:,-1]

from sklearn.model_selection import validation_curve

p_range=np.arange(10,1000,50)

train_score,test_score=validation_curve(RandomForestClassifier(random_state=42),
                                        X,Y,
                                        param_name="n_estimators",cv=5,
                                        param_range=p_range,scoring="accuracy",n_jobs=-1)
train_mean=np.mean(train_score,axis=1)
test_mean=np.mean(test_score,axis=1)

plt.plot(p_range,train_mean,label="train score",color='Black')
plt.plot(p_range,test_mean,label="cross validation score",color='Dimgrey')

plt.title('validation curve')
plt.xlabel('no. of tree')
plt.ylabel('Accuracy')
plt.legend(loc='Best')
plt.show()
```

```python
from sklearn.ensemble import ExtraTreesClassifier
ETF=ExtraTreesClassifier(n_estimators=100).fit(X,Y)
```

```python
feature_imp=ETF.feature_importances_
std=np.std([i.feature_importances_ for i in ETF],axis=0)
sns.barplot(std,X.columns)
plt.title('feature selection of tree')
plt.xlabel('feature importance')
plt.ylabel('feature columns')
pd.DataFrame({'Columns':X.columns,'Feature Importance':std})
```

```python
train_score,test_score=validation_curve(RandomForestClassifier(n_estimators=100,random_state=0),
                           X,Y,
                           param_name='max_depth',
                           param_range=np.arange(5,15,2),cv=5,
                           scoring='accuracy',
                           n_jobs=-1)

train_mean=np.mean(train_score,axis=0)
test_mean=np.mean(test_score,axis=0)

plt.plot(param_range,train_mean,label='train score',color='Black')
plt.plot(param_range,test_mean,label='cross validation score',color='Red')

plt.title('Max depth of random forest')
plt.xlabel('Max depth')
plt.ylabel('accuracy')
plt.legend(loc='Best')
```

```python
rfc=RandomForestClassifier(n_estimators=100,max_depth=9,random_state=0)
select=RFECV(rfc,step=1,cv=5).fit(X,Y)
```

```python
pd.DataFrame({'columns':X.columns,'ranking':select.ranking_,'support':select.support_})
```

```python
rfc.fit(X,Y)
rfc.score(X,Y)
```

```python
from sklearn.metrics import accuracy_score,confusion_matrix
y_predict=rfc.predict(test.iloc[:,:-1])
print('accuracy of rfe model',accuracy_score(y_predict,test.iloc[:,-1]))
confusion_matrix(test.iloc[:,-1],y_predict)
```
